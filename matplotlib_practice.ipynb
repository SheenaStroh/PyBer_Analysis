```python
%matplotlib inline
```

```python
# Import Dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot.
plt.plot(x_axis, y_axis)
```

```python
# Create the plot and add a label for the legend.
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis, marker="D", color="green", linewidth=2, label="Boston")

# Create labels for the x and y axes.
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")

# Set the y limit between 0 and 45.
ax.set_ylim(0,45)

# Create a title.
ax.set_title("PyBer Fare by Month")

# Add a grid
ax.grid()

# Add a legend
ax.legend()
```

```python
# Create Bar chart
plt.bar(x_axis,y_axis)
```

```python
# Create the plot
plt.bar(x_axis,y_axis, color="green", label='Boston')

# Create labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Create a title
plt.title("PyBer Fare by Month")

# Add the legend
plt.legend()
```

```python
plt.barh(x_axis,y_axis, color="magenta", label='Boston')
plt.xlabel("Fare($)")
plt.ylabel("Month")
plt.title("PyBer Fare by Month")
plt.gca().invert_yaxis()
plt.legend()
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color="cyan", label='Chicaco')
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")
ax.set_title("PyBer Fare by Month")
ax.invert_yaxis()
ax.legend()
```

```python
plt.plot(x_axis,y_axis, 'o')
```

```python
plt.scatter(y_axis, x_axis, color="red", label='Chicago')
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.title("PyBer Fare by Month")
plt.legend()
plt.xlim(0,50)
plt.gca().invert_yaxis()
```

```python
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
plt.scatter(x_axis, y_axis, s=[i*3 for i in y_axis])
```

```python
fig, ax = plt.subplots()
ax.scatter(y_axis,x_axis, color="skyblue", s=[i*5 for i in y_axis], alpha=0.8, linewidth=2, edgecolors="black", label='Boston')
ax.legend()
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")
ax.set_xlim(0,50)
ax.invert_yaxis()
```

```python
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
plt.subplots(figsize=(8,8))
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
plt.show()
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,8))
ax.pie(y_axis,explode=explode_values, colors=colors, labels=x_axis, autopct='%.1f%%', shadow=True, startangle=270)
ax.invert_yaxis()
plt.show()
```
